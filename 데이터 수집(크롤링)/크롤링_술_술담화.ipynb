```python
#네종류 전부
dict={}
alcohol_count=0
driver.get('https://www.sooldamhwa.com/damhwaMarket/listing/takju')
time.sleep(5)  # 2초간 정지
for type in range(1,5):
    time.sleep(1)
    driver.find_element_by_xpath(f'//*[@id="root"]/div[2]/div[2]/ul/li[{type}]').click()
    time.sleep(2)
    body= driver.find_element_by_tag_name('body')
    for i in range(30):
        body.send_keys(Keys.PAGE_DOWN)
        time.sleep(0.5)
    # 주류 개수
    time.sleep(2)
    num_list = driver.find_element_by_xpath('//*[@id="root"]/div[2]/div[3]/div/div[3]/div[1]/div/div/div[1]/div[1]').text
    prodcut_list_num = int(num_list)
    for product in tqdm_notebook(range(1,prodcut_list_num+1)):
        try:
            time.sleep(2)
            driver.find_element_by_xpath(f'//*[@id="root"]/div[2]/div[3]/div/div[3]/div[2]/div/div/div/div[{product}]/div/a/div[1]/div/div').click()
            time.sleep(1)
        except:
            continue
        data_info={}
    
        title=driver.find_element_by_xpath('//*[@id="root"]/div[2]/div[2]/div/div[1]/div[1]/div[2]/div/div[1]/div').text
        data_info['title']=title
    
        alcohol_type=driver.find_element_by_xpath('//*[@id="root"]/div[2]/div[2]/div/div[1]/div[1]/div[2]/div/div[3]/div[1]/span').text
        data_info['alcohol_type']=alcohol_type

        alcohol_degree=driver.find_element_by_xpath('//*[@id="root"]/div[2]/div[2]/div/div[1]/div[1]/div[2]/div/div[3]/div[2]/span').text
        data_info['alcohol_degree']=alcohol_degree

        capacity = driver.find_element_by_xpath('//*[@id="root"]/div[2]/div[2]/div/div[1]/div[1]/div[2]/div/div[3]/div[3]/span').text
        data_info['용량']=capacity
        
        price = driver.find_element_by_xpath('//*[@id="root"]/div[2]/div[2]/div/div[1]/div[1]/div[2]/div/div[4]/div[2]/div/span').text
        data_info['price']=price
        
        food_list=driver.find_elements_by_css_selector('.food-wrapper')
        foods=[]
        for food in food_list:
            foods.append(food.text)
        data_info['안주']=foods
 
        for x in range(1,6):
            flavor=driver.find_element_by_xpath(f'//*[@id="root"]/div[2]/div[2]/div/div[1]/div[4]/div/div[{x}]/div[2]').text
            data_info[flavor]=driver.find_element_by_xpath(f'//*[@id="root"]/div[2]/div[2]/div/div[1]/div[4]/div/div[{x}]/div[1]').text
        time.sleep(1)
        dict[alcohol_count]=data_info
        alcohol_count += 1
        time.sleep(1)
        driver.back()
        time.sleep(3)
```

```python
#한종류
alcohol_count=0
dict={}
num_list = driver.find_element_by_xpath('//*[@id="root"]/div[2]/div[3]/div/div[3]/div[1]/div/div/div[1]/div[1]').text
prodcut_list_num = int(num_list)
for product in tqdm_notebook(range(1,prodcut_list_num+1)):
    try:
        driver.find_element_by_xpath(f'//*[@id="root"]/div[2]/div[3]/div/div[3]/div[2]/div/div/div/div[{product}]/div/a/div[1]/div/div').click()

        data_info={}
        try:
            title=driver.find_element_by_xpath('//*[@id="root"]/div[2]/div[2]/div/div[1]/div[1]/div[2]/div/div[1]/div').text
            data_info['title']=title
        except:
            continue
        try:
            alcohol_type=driver.find_element_by_xpath('//*[@id="root"]/div[2]/div[2]/div/div[1]/div[1]/div[2]/div/div[3]/div[1]/span').text
            data_info['alcohol_type']=alcohol_type

            alcohol_degree=driver.find_element_by_xpath('//*[@id="root"]/div[2]/div[2]/div/div[1]/div[1]/div[2]/div/div[3]/div[2]/span').text
            data_info['alcohol_degree']=alcohol_degree

            capacity = driver.find_element_by_xpath('//*[@id="root"]/div[2]/div[2]/div/div[1]/div[1]/div[2]/div/div[3]/div[3]/span').text
            data_info['용량']=capacity
        except:
            continue
        
        try:
            food_list=driver.find_elements_by_css_selector('.food-wrapper')
            foods=[]
            for food in food_list:
                foods.append(food.text)
            data_info['안주']=foods
        except:
            continue
        try:
            for x in range(1,6):
                flavor=driver.find_element_by_xpath(f'//*[@id="root"]/div[2]/div[2]/div/div[1]/div[4]/div/div[{x}]/div[2]').text
                data_info[flavor]=driver.find_element_by_xpath(f'//*[@id="root"]/div[2]/div[2]/div/div[1]/div[4]/div/div[{x}]/div[1]').text
        except:
            continue
        dict[alcohol_count]=data_info
        alcohol_count += 1
        time.sleep(1)
        driver.back()
    except:
        continue
    time.sleep(3)
```
