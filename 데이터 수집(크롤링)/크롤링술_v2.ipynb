```python
import pandas as pd  # 판다스 : 데이터분석 라이브러리
import numpy as np   # 넘파이 : 숫자, 행렬 데이터 라이브러리
import re
import chromedriver_autoinstaller
from bs4 import BeautifulSoup    # html 데이터를 전처리
from selenium import webdriver   # 웹 브라우저 자동화
from selenium.webdriver.common.keys import Keys
import time    # 서버와 통신할 때 중간중간 시간 지연. 보통은 1초
from tqdm import tqdm_notebook   # for문 돌릴 때 진행상황을 %게이지로 알려준다.

# 워닝 무시
import warnings
warnings.filterwarnings('ignore')
```

```python
options = webdriver.ChromeOptions()
chrome_path = chromedriver_autoinstaller.install()
driver = webdriver.Chrome(chrome_path, options=options)
```

```python
pd.set_option('display.width', 500)
# row 생략 없이 출력
pd.set_option('display.max_rows', 1000)
# col 생략 없이 출력
pd.set_option('display.max_columns', 1000)
```

```python
index_count=1
count = 7
for next in tqdm_notebook(range(1,36)):
    for x in range(1,16):
        data_info={}
        for i in range(1,6):
            data_info[i]=driver.find_element_by_xpath(f'//*[@id="content"]/div[4]/ul/li[{x}]/div[2]/div[2]/div/span[{i}]/em').text
        data_info['img_src']=driver.find_element_by_xpath('//*[@id="content"]/div[4]/ul/li[1]/div[1]/div[1]/a[1]/img').get_attribute('src')
        data_info['감정표현']=driver.find_element_by_xpath(f'//*[@id="content"]/div[4]/ul/li[{x}]/div[2]/div[3]/div/span[2]/div/a/span[3]').text
        data_info['조회수']=driver.find_element_by_xpath(f'//*[@id="content"]/div[4]/ul/li[{x}]/div[2]/div[3]/div/span[3]/em').text
        dict[index_count]=data_info
        index_count= index_count+1
    if next<6:
        driver.find_element_by_xpath(f'//*[@id="paginate"]/a[{next}]').click()
    elif next<34:
        driver.find_element_by_xpath(f'//*[@id="paginate"]/a[6]').click()
    else:
        driver.find_element_by_xpath(f'//*[@id="paginate"]/a[{count}]').click()
        count=count+1
      
```
