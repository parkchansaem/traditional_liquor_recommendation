```python
takju= pd.read_excel('takju.xlsx')
gwasilju= pd.read_excel('gwasilju.xlsx')
jeunglyuju= pd.read_excel('jeunglyuju.xlsx')
yag_cheongju= pd.read_excel('yag_cheongju.xlsx')
```

```python
takju['section']='takju'
gwasilju['section']='gwasilju'
jeunglyuju['section']='jeunglyuju'
yag_cheongju['section']='yag_cheongju'
```

```python
df= pd.concat([takju,gwasilju,jeunglyuju,yag_cheongju],ignore_index=True,sort=False)
df
```

```python
# 중복값 제거
df=df.drop_duplicates(['name'],keep='first',ignore_index=True)
df
```

데이터 전처리

```python
# 불필요한 문자제거
df['food_name']=df['food_name'].str.replace("'","")
df
```

```python
# 리스트화 
for i in range(len(df)):
    df['food_name'][i]=df['food_name'][i].replace("[","").replace("]","").replace(" ","").split(",")
```

```python
# 특정 단어 찾기 
a_list=[]
food_name_input=input()
for i in range(len(df)):
    for j in df['food_name'][i]:
        if j==food_name_input:
            a_list.append(i)
check_list = df.loc[a_list]
```

```python
check_list
```

```python
#섹션별 한종류만
check_list=check_list.drop_duplicates(subset='section')
check_list.name
for a in check_list.name:
    print(a)
```

```python
# 어울리는 안주를 CountVectorizer를 위해 컬럼 추가
df['food_name_literal'] = df['food_name'].apply(lambda x : (' ').join(x))
count_vect = CountVectorizer(min_df=0, ngram_range=(1,3))
food_mat = count_vect.fit_transform(df['food_name_literal'])
food_sim = cosine_similarity(food_mat, food_mat)
food_sim_sorted_ind =food_sim.argsort()[:,::-1]
print(food_sim_sorted_ind[:5])
```

```python
# 코사인 유사도를 기준으로 역순으로 정렬해주는 함수 구현
def find_sim_food_ver1(df,sorted_ind,title_food, top_n=10):
    title = df[df['name']== title_food]
    
    title_index = title.index.values
    similar_indexes = sorted_ind[title_index,:(top_n)]
    
    print(similar_indexes)
    similar_indexes = similar_indexes.reshape(-1)
    
    return df.iloc[similar_indexes]
```

```python
# 4가지 종류 추천 알고리즘 실행
dict = {}
count_num = 0
for i in check_list.name:
    dict[count_num]=find_sim_food_ver1(df,food_sim_sorted_ind,i,9)
    count_num += 1
    
```

```python
#데이터프레임 합치기
df_list = []
for r_list in range(len(dict)):
    globals()[f'df_{r_list}'] = dict[r_list]
    df_list.append(globals()[f'df_{r_list}'])
```

```python
#데이터프레임 합치기2
df_df = pd.concat(df_list,ignore_index=True,sort=False)
```

```python
#합친 데이터 중복 제거
df_df=df_df.drop_duplicates(['name'],keep='first',ignore_index=True)
```
